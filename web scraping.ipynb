# Web Scraping
## Importing Libraries

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
```

## Sent HTTP requests to the target website

```python
webpage = requests.get('https://quotes.toscrape.com/').text
```

```python
webpage
```

## Parsed HTML content using BeautifulSoup

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('h1')[0].text
```

## Extracted relevant data fields

```python
content = soup.find_all('div',class_='quote')
content
```

## Cleaned and structured the data into a DataFrame

```python
quotes=[]
author=[]
tags=[]
for i in content:
    quotes.append(i.find('span',class_='text').text.strip())
    author.append(i.find('small',class_='author').text.strip())
    t = i.find('div',class_='tags')
    tags.append(", ".join(a.text.strip() for a in t.find_all("a", class_="tag")))
    
```

```python
df = pd.DataFrame({'quotes':quotes,'author':author,'tags':tags})
```
